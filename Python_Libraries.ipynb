# Numpy - Numerical Python


Numpy Array

```python
import numpy as np

a = np.array([1,2,3,4])
print(a)
```

```python
a.shape
```

```python
b = np.array([(1,2,3,4),(5,6,7,8)])
print(b)
```

```python
b.shape
```

```python
c =  np.array([(1,2,3,4),(5,6,7,8)], dtype = float)
print(c)
```

Intial placeholders in numpy array

```python
#create a numpy array of zeros
x = np.zeros((4,4))
print(x)
```

```python
#create a numpy array of ones
y = np.ones((4,5))
print(y)
```

```python
#create an array of a particular value
z = np.full((5,4),5)
print(z)
```

```python
#create an identity matrix
s = np.eye(5)
print(s)
```

```python
#array with random values
a = np.random.random((5,5))
print(a)
```

```python
#array with random integer values within a specific range
b = np.random.randint(10,100,(3,5))
print(b)
```

```python
# array of evenly spaced values ---> specifying the number of values required
d = np.linspace(10,30,6)
print(d)
```

```python
#array of evenly spaced values ---> specifying the steps
e = np.arange(10,30,5)
print(e)
```

```python
# convert a list to numpy array
list1 = [10,20,30,30]
np_array = np.asarray(list1)
print(np_array)
type(np_array)
```

Array Manipulation

```python
# Transpose of an array

a = np.random.randint(10,20,(3,5))
print(a)
```

```python
trans = np.transpose(a)
print(trans)
print(trans.shape)
```

```python
# Another method to find Transpose of an array

a = np.random.randint(10,20,(3,5))
print(a)
```

```python
trans = a.T
print(trans)
print(trans.shape)
```

```python
#Reshaping a array
a = np.random.randint(0,10,(2,3))
print(a)

b = a.reshape(3,2)
print(b)
```

Mathematical operations on a numpy array

```python
# Addition Subtraction Division Multiplication
a = np.random.randint(10,20,(3,3))
b = np.random.randint(30,50,(3,3))

print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

# Matplotlib


Useful for plots

```python
#importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#importing numpy library
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Ploting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#adding title x-axis & z-axis labels
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('cosine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10 , 10, 20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-5,5, 50)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'b--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Hindi', 'Latin', 'Spanish', 'Russian']
people = [100, 200, 50, 60, 120]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie chart


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Hindi', 'Latin', 'Spanish', 'Russian']
people = [100, 200, 50, 60, 120]
ax.pie(people, labels = languages , autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,0.5,0.5])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c=z , cmap= 'Blues')
plt.show()
```
